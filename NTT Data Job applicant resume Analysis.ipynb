```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df1=pd.read_csv('RESUMEFILES.csv')

df1
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(x='College Branch', data=df1)
```

```python
#Max applicants in ECE!

sns.barplot(x='Pay Grade', y='Company Tier', data=df1)
```

```python
sns.barplot(x='Pay Grade', y='Applied for job level', data=df1)

#generally speaking, candidates from good companies were offered higher pay grade. But a lot of candidates from good companies were rejected!
#this could be because their performance in prev company wasn't upto par, or because they were applying for higher pay grade than what they were qualified for!
```

```python
sns.barplot(x='Pay Grade', y='Number of skills', data=df1)

#even tho b tier candidates have skills <= C tier and rejected candidates, they have higher pay!
```

```python
sns.barplot(x='Pay Grade', y='Years of Exp', data=df1)

#A lot of candidates despite having yrs of exp>6, were rejected. but by and large yrs of exp may play a pretty big determining role
```

```python
sns.barplot(x='Applied for job level', y='Years of Exp',data=df1)
```

```python
sns.barplot(x='Applied for job level', y='Years of Exp',data=df1,hue='Selection Status')

#as we can see yrs of exp play a more significant role in higher grade jobs (8 and above)
#many candidates applying for grade 7 jobs and lower, despite having higher yrs of exp were rejected
#this could be due to the fact that they had better quality skills suited for the job which they were able to perfect in shorter quantity of time
```
